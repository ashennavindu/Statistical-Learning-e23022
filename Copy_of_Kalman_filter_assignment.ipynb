# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

# Solution

## 1. Show that

$$
x_k^- \sim \mathscr{N}(m_k^-, P_k^-),
$$

where

$$
m_k^- = A_{k-1}m_{k-1},
$$

and

$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

### Solution

The prediction equation is

$$
x_k^- = A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1}.
$$

Since both $x_{k-1}^+$ and $w_{k-1}$ are Gaussian random variables, and a linear combination of Gaussian random variables is also Gaussian,

$$
x_k^- \sim \mathscr{N}(m_k^-, P_k^-).
$$

### Mean

Taking the expectation,

$$
\begin{aligned}
m_k^-
&= E[x_k^-] \\
&= E[A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1}] \\
&= A_{k-1}E[x_{k-1}^+] + G_{k-1}E[w_{k-1}] \\
&= A_{k-1}m_{k-1},
\end{aligned}
$$

because

$$
E[w_{k-1}] = 0.
$$

### Covariance

The covariance is

$$
\begin{aligned}
P_k^-
&= \mathrm{Var}(x_k^-) \\
&= \mathrm{Var}(A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1}) \\
&= A_{k-1}\mathrm{Var}(x_{k-1}^+)A_{k-1}^T
   + G_{k-1}\mathrm{Var}(w_{k-1})G_{k-1}^T \\
&= A_{k-1}P_{k-1}A_{k-1}^T
   + G_{k-1}\Sigma_p G_{k-1}^T.
\end{aligned}
$$

Therefore,

$$
\boxed{
x_k^- \sim \mathscr{N}(m_k^-, P_k^-)
}
$$

where

$$
\boxed{
m_k^- = A_{k-1}m_{k-1}
}
$$

and

$$
\boxed{
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
}
$$

---

## 2. Show that

$$
y_k^- \sim \mathscr{N}(H_km_k^-,\, H_kP_k^-H_k^T+\Sigma_m).
$$

### Solution

The measurement equation is

$$
y_k^- = H_kx_k^- + z_k.
$$

Since $x_k^-$ and $z_k$ are Gaussian random variables, $y_k^-$ is also Gaussian.

### Mean

Taking the expectation,

$$
\begin{aligned}
E[y_k^-]
&= E[H_kx_k^- + z_k] \\
&= H_kE[x_k^-] + E[z_k] \\
&= H_km_k^-,
\end{aligned}
$$

because

$$
E[z_k]=0.
$$

### Covariance

The covariance is

$$
\begin{aligned}
\mathrm{Var}(y_k^-)
&= \mathrm{Var}(H_kx_k^- + z_k) \\
&= H_k\mathrm{Var}(x_k^-)H_k^T + \mathrm{Var}(z_k) \\
&= H_kP_k^-H_k^T + \Sigma_m.
\end{aligned}
$$

Therefore,

$$
\boxed{
y_k^- \sim
\mathscr{N}(H_km_k^-,
H_kP_k^-H_k^T+\Sigma_m)
}
$$

---

## 3. Show that

$$
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\sim
\mathscr{N}
\left(
\begin{bmatrix}
m_k^-\\
H_km_k^-
\end{bmatrix},
\begin{bmatrix}
P_k^- & P_k^-H_k^T\\
H_kP_k^- & H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}
\right).
$$

### Solution

The joint mean vector is

$$
\begin{aligned}
E
\left[
\begin{pmatrix}
x_k^-\\
y_k^-
\end{pmatrix}
\right]
=
\begin{pmatrix}
E[x_k^-]\\
E[y_k^-]
\end{pmatrix}
=
\begin{pmatrix}
m_k^-\\
H_km_k^-
\end{pmatrix}.
\end{aligned}
$$

Next, compute the cross-covariance.

$$
\begin{aligned}
\mathrm{Cov}(x_k^-,y_k^-)
&=
\mathrm{Cov}(x_k^-,
H_kx_k^-+z_k)\\
&=
\mathrm{Cov}(x_k^-,H_kx_k^-)
+
\mathrm{Cov}(x_k^-,z_k).
\end{aligned}
$$

Since the measurement noise is independent of the state,

$$
\mathrm{Cov}(x_k^-,z_k)=0.
$$

Therefore,

$$
\boxed{
\mathrm{Cov}(x_k^-,y_k^-)
=
P_k^-H_k^T.
}
$$

Similarly,

$$
\mathrm{Cov}(y_k^-,x_k^-)
=
H_kP_k^-.
$$

The covariance of $y_k^-$ is

$$
H_kP_k^-H_k^T+\Sigma_m.
$$

Hence, the joint covariance matrix becomes

$$
\begin{bmatrix}
P_k^- & P_k^-H_k^T\\
H_kP_k^- & H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}.
$$

Therefore,

$$
\boxed{
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\sim
\mathscr{N}
\left(
\begin{bmatrix}
m_k^-\\
H_km_k^-
\end{bmatrix},
\begin{bmatrix}
P_k^- & P_k^-H_k^T\\
H_kP_k^- & H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}
\right).
}
$$

---

## 4. Show that

$$
x_k^+ \equiv (x_k^- \mid y_k^- = y_k^{obs})
\sim
\mathscr{N}(m_k,P_k),
$$

where

$$
K_k
=
P_k^-H_k^T
\left(H_kP_k^-H_k^T+\Sigma_m\right)^{-1},
$$

$$
m_k
=
m_k^-
+
K_k
\left(
y_k^{obs}-H_km_k^-
\right),
$$

and

$$
P_k
=
(I-K_kH_k)P_k^-.
$$

### Solution

From Question 3, the joint distribution of the predicted state and predicted measurement is

$$
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\sim
\mathscr{N}
\left(
\begin{bmatrix}
m_k^-\\
H_km_k^-
\end{bmatrix},
\begin{bmatrix}
P_k^- & P_k^-H_k^T\\
H_kP_k^- & H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}
\right).
$$

For a jointly Gaussian random vector, the conditional distribution is also Gaussian.

The Kalman Gain is

$$
\boxed{
K_k
=
P_k^-H_k^T
\left(H_kP_k^-H_k^T+\Sigma_m\right)^{-1}.
}
$$

The conditional mean (updated state estimate) is

$$
\begin{aligned}
m_k
&=
m_k^-
+
K_k
\left(
y_k^{obs}-H_km_k^-
\right).
\end{aligned}
$$

The conditional covariance is

$$
\begin{aligned}
P_k
&=
P_k^-
-
P_k^-H_k^T
\left(H_kP_k^-H_k^T+\Sigma_m\right)^{-1}
H_kP_k^-\\
&=
(I-K_kH_k)P_k^-.
\end{aligned}
$$

Therefore,

$$
\boxed{
x_k^+
=
(x_k^-|y_k^-=y_k^{obs})
\sim
\mathscr{N}(m_k,P_k).
}
$$

---

## 5. Find

$$
E[x_k^-|y_k^-=y_k^{obs}]
$$

and

$$
\mathrm{Var}(x_k^-|y_k^-=y_k^{obs}).
$$

### Solution

From the conditional Gaussian distribution obtained in Question 4,

the conditional expectation is

$$
\begin{aligned}
E[x_k^-|y_k^-=y_k^{obs}]
&=
m_k^-+
K_k
\left(
y_k^{obs}-H_km_k^-
\right).
\end{aligned}
$$

Hence,

$$
\boxed{
E[x_k^-|y_k^-=y_k^{obs}]
=
m_k.
}
$$

Similarly, the conditional covariance is

$$
\begin{aligned}
\mathrm{Var}(x_k^-|y_k^-=y_k^{obs})
&=
P_k^--
P_k^-H_k^T
\left(H_kP_k^-H_k^T+\Sigma_m\right)^{-1}
H_kP_k^-\\
&=
(I-K_kH_k)P_k^-.
\end{aligned}
$$

Therefore,

$$
\boxed{
\mathrm{Var}(x_k^-|y_k^-=y_k^{obs})
=
P_k.
}
$$

Thus,

$$
\boxed{
E[x_k^-|y_k^-=y_k^{obs}]
=
m_k
}
$$

and

$$
\boxed{
\mathrm{Var}(x_k^-|y_k^-=y_k^{obs})
=
P_k.
}
$$

# Q. 1-D Example

Consider the scalar linear-Gaussian filter model

$$
x_k^- = ax_{k-1}^+ + w_{k-1},
\qquad
w_{k-1}\sim\mathscr{N}(0,q),
$$

$$
y_k^- = hx_k^- + z_k,
\qquad
z_k\sim\mathscr{N}(0,r),
$$

with prior

$$
x_0\sim\mathscr{N}(m_0,P_0).
$$

Define

$$
Y_k=\{y_1^{obs},y_2^{obs},\ldots,y_k^{obs}\}.
$$

---

## 1. Show that

$$
m_k^- = am_{k-1},
$$

and

$$
P_k^- = a^2P_{k-1}+q.
$$

### Solution

The prediction equation is

$$
x_k^- = ax_{k-1}^+ + w_{k-1}.
$$

Since both $x_{k-1}^+$ and $w_{k-1}$ are Gaussian random variables, $x_k^-$ is also Gaussian.

### Mean

Taking the expectation,

$$
\begin{aligned}
m_k^-
&=E[x_k^-] \\
&=E[ax_{k-1}^+ + w_{k-1}]\\
&=aE[x_{k-1}^+] + E[w_{k-1}]\\
&=am_{k-1},
\end{aligned}
$$

because

$$
E[w_{k-1}]=0.
$$

### Variance

The predicted variance is

$$
\begin{aligned}
P_k^-
&=\mathrm{Var}(ax_{k-1}^+ + w_{k-1})\\
&=a^2\mathrm{Var}(x_{k-1}^+) + \mathrm{Var}(w_{k-1})\\
&=a^2P_{k-1}+q.
\end{aligned}
$$

Therefore,

$$
\boxed{
m_k^- = am_{k-1},
\qquad
P_k^- = a^2P_{k-1}+q.
}
$$

---

## 2. Show that

$$
m_k
=
m_k^-+K_kv_k
=
m_k^-+\frac{P_k^-h}{S_k}(y_k^{obs}-hm_k^-),
$$

and

$$
P_k
=
(1-K_kh)P_k^-
=
\left(1-\frac{P_k^-h^2}{S_k}\right)P_k^-.
$$

### Solution

The innovation (measurement residual) is

$$
\boxed{
v_k
=
y_k^{obs}-hm_k^-.
}
$$

The innovation variance is

$$
\boxed{
S_k=h^2P_k^-+r.
}
$$

The Kalman Gain is

$$
\boxed{
K_k=\frac{P_k^-h}{S_k}.
}
$$

The updated mean is

$$
\begin{aligned}
m_k
&=
m_k^-+K_kv_k\\
&=
m_k^-+\frac{P_k^-h}{S_k}(y_k^{obs}-hm_k^-).
\end{aligned}
$$

The updated variance is

$$
\begin{aligned}
P_k
&=(1-K_kh)P_k^-\\
&=
\left(1-\frac{P_k^-h^2}{S_k}\right)P_k^-.
\end{aligned}
$$

Therefore,

$$
\boxed{
m_k
=
m_k^-+\frac{P_k^-h}{S_k}(y_k^{obs}-hm_k^-)
}
$$

and

$$
\boxed{
P_k
=
\left(1-\frac{P_k^-h^2}{S_k}\right)P_k^-.
}
$$

---

## 3. Show that the predictive measurement distribution (before observing $y_k$) is

$$
p(y_k^-|Y_{k-1})
=
\mathscr{N}(hm_k^-,\,h^2P_k^-+r).
$$

### Solution

The measurement equation is

$$
y_k^- = hx_k^-+z_k.
$$

Since

$$
x_k^-\sim\mathscr{N}(m_k^-,P_k^-)
$$

and

$$
z_k\sim\mathscr{N}(0,r),
$$

their linear combination is also Gaussian.

### Mean

$$
\begin{aligned}
E[y_k^-]
&=
hE[x_k^-]+E[z_k]\\
&=
hm_k^-.
\end{aligned}
$$

### Variance

$$
\begin{aligned}
\mathrm{Var}(y_k^-)
&=
h^2P_k^-+r.
\end{aligned}
$$

Therefore,

$$
\boxed{
p(y_k^-|Y_{k-1})
=
\mathscr{N}(hm_k^-,
h^2P_k^-+r).
}
$$

---

## 4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$) is

$$
p(y_k^-|Y_k)
=
\mathscr{N}(hm_k,h^2P_k+r).
$$

### Solution

After incorporating the measurement, the posterior state distribution becomes

$$
x_k^+\sim\mathscr{N}(m_k,P_k).
$$

Using the measurement equation

$$
y_k^- = hx_k^+ + z_k,
$$

the posterior-predictive measurement distribution is also Gaussian.

### Mean

$$
\begin{aligned}
E[y_k^-|Y_k]
&=
hE[x_k^+]\\
&=
hm_k.
\end{aligned}
$$

### Variance

$$
\begin{aligned}
\mathrm{Var}(y_k^-|Y_k)
&=
h^2P_k+r.
\end{aligned}
$$

Therefore,

$$
\boxed{
p(y_k^-|Y_k)
=
\mathscr{N}(hm_k,
h^2P_k+r).
}
$$

$$
m_1^-
=
am_0
=
1(0)
=

P_1^-
=
a^2P_0+q
=
1^2(1)+0.2

=

## Part B

Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Simulation Parameters
# -----------------------------
dt = 1.0

A = np.array([[1,0,dt,0],
              [0,1,0,dt],
              [0,0,1,0],
              [0,0,0,1]])

H = np.array([[1,0,0,0],
              [0,1,0,0]])

G = np.array([[0.5*dt**2,0],
              [0,0.5*dt**2],
              [dt,0],
              [0,dt]])

Q = 0.1*np.eye(2)
R = 4*np.eye(2)

# Initial estimate
x = np.array([0,0,1,0.5],dtype=float)

P = np.eye(4)

# -----------------------------
# Generate true trajectory
# -----------------------------
N = 50

true_states = []
gps_measurements = []
filtered_states = []

true = np.array([0,0,1,0.5],dtype=float)

for k in range(N):

    # True motion
    true = A @ true
    true_states.append(true.copy())

    # Noisy GPS measurement
    z = H @ true + np.random.multivariate_normal([0,0],R)
    gps_measurements.append(z)

    # -----------------------------
    # Prediction
    # -----------------------------
    x_pred = A @ x
    P_pred = A @ P @ A.T + G @ Q @ G.T

    # -----------------------------
    # Kalman Gain
    # -----------------------------
    S = H @ P_pred @ H.T + R

    K = P_pred @ H.T @ np.linalg.inv(S)

    # -----------------------------
    # Update
    # -----------------------------
    x = x_pred + K @ (z - H @ x_pred)

    P = (np.eye(4) - K @ H) @ P_pred

    filtered_states.append(x.copy())

true_states = np.array(true_states)
gps_measurements = np.array(gps_measurements)
filtered_states = np.array(filtered_states)

# -----------------------------
# Plot Results
# -----------------------------
plt.figure(figsize=(8,6))

plt.plot(true_states[:,0],
         true_states[:,1],
         label="True Path",
         linewidth=2)

plt.scatter(gps_measurements[:,0],
            gps_measurements[:,1],
            color="red",
            s=20,
            label="GPS Measurements")

plt.plot(filtered_states[:,0],
         filtered_states[:,1],
         linewidth=2,
         label="Kalman Filter")

plt.xlabel("x Position")
plt.ylabel("y Position")
plt.title("2D Position Estimation using Kalman Filter")

plt.legend()
plt.grid(True)
plt.show()
```
