## Bertrand's Paradox: Random Chord Probability

The problem asks for the probability that a randomly chosen chord of a circle has a length greater than the side length of an inscribed equilateral triangle. We will demonstrate that the answer depends on how we define "random chord."

First, let's establish the geometric condition. Consider a circle with radius `R`.

The side length `s` of an equilateral triangle inscribed in a circle of radius `R` is given by `s = R * sqrt(3)`.

Now, consider a chord. Its length `L` is determined by its distance `d` from the center of the circle, using the formula `L = 2 * sqrt(R^2 - d^2)`.

We want to find the condition where `L > s`:

```
2 * sqrt(R^2 - d^2) > R * sqrt(3)
4 * (R^2 - d^2) > 3 * R^2
4R^2 - 4d^2 > 3R^2
R^2 > 4d^2
d^2 < R^2 / 4
|d| < R / 2
```

Thus, a chord's length will be greater than the side length of an inscribed equilateral triangle if and only if its distance `d` from the center of the circle is less than `R/2`.

````python
import numpy as np
import matplotlib.pyplot as plt

# Set a fixed random seed for reproducibility
np.random.seed(42)

# Define the circle radius
R = 1

# Calculate the side length of an inscribed equilateral triangle
s_equilateral = R * np.sqrt(3)
print(f"Radius of the circle (R): {R}")
print(f"Side length of inscribed equilateral triangle: {s_equilateral:.4f}")

# The condition for a chord length L > s_equilateral is d < R/2
d_threshold = R / 2
print(f"Distance from center threshold (d < R/2): {d_threshold:.4f}")
````

### Model 1: Choosing Two Random Endpoints on the Circumference

In this model, we choose two points independently and uniformly at random on the circumference of the circle. The chord is the line segment connecting these two points.

Let the first point be fixed at $(R, 0)$ without loss of generality. The second point is chosen by picking an angle `theta` uniformly from `[0, 2π)`. The length of the chord connecting $(R, 0)$ and $(R\cos\theta, R\sin\theta)$ is given by $L = 2R \sin(\theta/2)$.

The condition $L > R\sqrt{3}$ becomes $2R \sin(\theta/2) > R\sqrt{3}$, which simplifies to $\sin(\theta/2) > \sqrt{3}/2$.

Since $\theta \in [0, 2\pi)$, $\theta/2 \in [0, \pi)$. In this range, $\sin(x) > \sqrt{3}/2$ when $x \in (\pi/3, 2\pi/3)$.

Therefore, we need $\pi/3 < \theta/2 < 2\pi/3$, which means $2\pi/3 < \theta < 4\pi/3$. The length of this favorable interval for $\theta$ is $4\pi/3 - 2\pi/3 = 2\pi/3$. The total range for $\theta$ is $2\pi$.

The theoretical probability for this model is $(2\pi/3) / (2\pi) = 1/3$.

````python
num_simulations = 100000
chords_exceeding_s_model1 = 0

for _ in range(num_simulations):
    # Choose two random angles for the endpoints (0 to 2*pi radians)
    theta1 = np.random.uniform(0, 2 * np.pi)
    theta2 = np.random.uniform(0, 2 * np.pi)

    # Calculate the angle between the two radii (can be simplified by fixing one point)
    # The length of the chord is 2R * sin(delta_theta / 2)
    delta_theta = abs(theta1 - theta2)
    if delta_theta > np.pi: # Ensure delta_theta is the smaller angle between the two radii
        delta_theta = 2 * np.pi - delta_theta

    chord_length = 2 * R * np.sin(delta_theta / 2)

    if chord_length > s_equilateral:
        chords_exceeding_s_model1 += 1

probability_model1 = chords_exceeding_s_model1 / num_simulations
print(f"\n--- Model 1: Random Endpoints ---")
print(f"Simulated probability: {probability_model1:.4f}")
print(f"Theoretical probability: {1/3:.4f}")
````

### Model 2: Choosing a Random Midpoint within the Circle

In this model, we choose a point uniformly at random *within* the circle. This point serves as the midpoint of the chord. The chord is then uniquely determined as the line segment perpendicular to the radius passing through this chosen midpoint.

The condition for the chord length to be greater than $s_{\text{equilateral}}$ is that the distance of the chord's midpoint from the center, `d`, must be less than $R/2$.

If we pick a random point (x, y) within the circle, its distance from the origin is $d = \sqrt{x^2 + y^2}$. We need this point to fall within a smaller concentric circle of radius $R/2$.

The probability is the ratio of the area of the inner circle (radius $R/2$) to the area of the outer circle (radius $R$).

Area of inner circle = $\pi (R/2)^2 = \pi R^2 / 4$
Area of outer circle = $\pi R^2$

The theoretical probability for this model is $(\pi R^2 / 4) / (\pi R^2) = 1/4$.

````python
num_simulations = 100000
chords_exceeding_s_model2 = 0

for _ in range(num_simulations):
    # Choose random coordinates (x, y) within the circle
    # Easier to pick a random radius r from [0, R] and angle phi from [0, 2*pi]
    r_midpoint = R * np.sqrt(np.random.uniform(0, 1)) # This correctly samples uniformly by area
    phi_midpoint = np.random.uniform(0, 2 * np.pi)

    # The distance 'd' of the midpoint from the center is r_midpoint
    d = r_midpoint

    if d < d_threshold:
        chords_exceeding_s_model2 += 1

probability_model2 = chords_exceeding_s_model2 / num_simulations
print(f"\n--- Model 2: Random Midpoint ---")
print(f"Simulated probability: {probability_model2:.4f}")
print(f"Theoretical probability: {1/4:.4f}")
````

### Conclusion

As demonstrated by the simulations and theoretical calculations:

*   **Model 1 (Random Endpoints):** The probability that the chord length exceeds the side length of the inscribed equilateral triangle is approximately **1/3**.
*   **Model 2 (Random Midpoint):** The probability that the chord length exceeds the side length of the inscribed equilateral triangle is approximately **1/4**.

These different results for the "same" problem illustrate Bertrand's Paradox. The core issue is that the phrase "a chord of a circle is chosen at random" is ambiguous. There is no single, unique way to define a uniform probability distribution over the set of all possible chords in a circle. Each model implies a different underlying sample space and thus leads to a different probability.
